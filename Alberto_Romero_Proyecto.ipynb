# Proyecto: Estudio de comportamiento en Spotify

El proyecto se basa en analizar el comportamiento de los usuarios de Spotify, estudiando datos reales.
La fuente de datos se encuentra en:

https://www.aicrowd.com/challenges/spotify-sequential-skip-prediction-challenge/dataset_files

1. Se necesita una cuenta en www.aicrowd.com, web de IA y data science. El registro es gratuito. Los datos son de un problema que se planteó previamente (y que se cerró en enero de 2019). Las puntuaciones y soluciones de los participantes no se hicieron públicas. En esa URL, vamos a _resources_ y tenemos todos los enlaces. 

2. Primero nos interesa Dataset Description (172 K), es un pdf que describe las variables (features). Vemos que existen dos ficheros, uno de sesiones y otro de canciones. Las sesiones tienen variables del comportamiento de un usuario dentro de su cuenta de Spotify, por ejemplo: id de sesión, si ha cambiado de contexto (si la ha cambiado de playlist, se llama  _context_switch_ ), si ha pausado entre canción y canción ( _no_pause_before_play_ ), etc.

El otro fichero es el de canciones ( _tracks_ ), que tiene variables propias de una canción: la duración, el año en el que se lanzó, etc.

3. Ahora el siguiente paso es descargar el dataset para analizarlo. Aquí hay varias opciones: tenemos unos enlaces para descargar las partes que nos interesan, otro enlace descarga todo el dataset de entrenamiento (56 GB) y otro enlace decarga un dataset de entrenamiento "mini" de 17,2 MB, para probar en local. Como vamos a hacer estas pruebas en local, elegimos el dataset "mini" de 17,2 MB, que es un muestreo del dataset grande.

En el enlace se descarga un .tar.gz que contiene a su vez 2 ficheros. En Windows recomendamos utilizar 7zip para descomprimirlos. Como ya hemos visto del punto 2, un fichero corresponde con las sesiones de usuarios y otro con un listado de canciones.

### 4. Variables dummies

Nos preparamos eliminando las variables categóricas de id que no queremos que se generen como dummies porque no entrarán como variables del modelo

#### Hay que entender que en el conjunto { 'not_skipped', 'skip_1', 'skip_2' ,'skip_3'} solo uno puede ser el target y es excluyente con los otros tres atributos, puesto que las variables son excluyentes entre sí (los estados de cuándo acaba de reproducirse la canción). Por esto, no se podrán usar skip_1, skip_2 y skip_3 como variables de entrada del algoritmo para predecir not_skipped (esos valores no están disponibles en la entrada porque son todos estados de salida)
Igualmente, eliminamos 'hist_user_behavior_reason_end' porque ese dato solo lo sabríamos al acabar la reproducción de la canción, cuando ya sabríamos si la canción se ha reproducido entera o no. El problema que queremos resolver es predecir, cuando una canción empieza (con sus valores propios de la canción y los valores propios de la sesión del usuario), si el usuario escuchará la canción entera.

```python
# Vemos que queremos eliminar los registros tipo id
# No es necesario hacer una copia del dataframe df_sesion_y_track_left porque el drop no lo modifica

df_sesion_y_track_sin_id = df_sesion_y_track_left.drop(['session_id', 'track_id_clean','track_id', 'skip_1', 'skip_2', 'skip_3', 'hist_user_behavior_reason_end' ], axis=1)
```

```python
# Comprobamos que el drop no ha modificado el dataframe y que los campos id y skip siguen en el dataframe
df_sesion_y_track_left.columns.isin(['session_id', 'track_id_clean','track_id','skip_1', 'skip_2', 'skip_3', 'hist_user_behavior_reason_end']).any()
```

```python
# Comprobamos que el drop sí ha modificado el dataframe y que los campos id y skip no están en el dataframe
df_sesion_y_track_sin_id.columns.isin(['session_id', 'track_id_clean','track_id', 'skip_1', 'skip_2', 'skip_3', 'hist_user_behavior_reason_end']).any()
```

### 5. Entrenamiento y test
Hasta ahora todas las transformaciones que hemos aplicado son externas al valor de los propios datos (podemos aplicarlas en otro dataset que contenga los mismos atributos sin conocer el valor de los atributos). En este paso separamos los datos de entrenamiento de los de test, porque los drops y dummies se pueden aplicar a un nuevo conjunto de test para ver las métricas de eficiencia en datos que no hayamos usado para entrenar ni validar

```python
# Recordamos que el target es 'not_skipped'

X = df_sesion_track_dummies_invariante.loc[:, df_sesion_track_dummies_invariante.columns != 'not_skipped']
y = df_sesion_track_dummies_invariante.loc[:,'not_skipped']

# Elegimos un 33% para test y la semilla de 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### 7. Balanceamos las clases

```python
# Posteriormente se incluirá la referencia a: 'TowardsDataScience: Having an imbalanced dataset? Here's is how you can fix it'
```

```python
# Como las clases no están balanceadas y tenemos muchas muestras, vamos a quedarnos 
# con un subconjunto de target = 0 (mayoritario) de la misma longitud que el target = 1
# para introducir un sesgo menor y además poder trabajar con la exactitud, que es una métrica
# que funciona bien cuando las clases están balanceadas

# Basándonos en el artículo de TowardsDataScience, vamos a crear una función que ejecute internamente el código
# propuesto, para poder generalizar en otro momento

def balancea(df,target,L):
    shuffled_df = df.sample (frac=1, random_state=4) # frac=1 es para cogerlo todo
    target_1_df = shuffled_df.loc[shuffled_df[target]==1] # Todo el target=1
    target_0_df = shuffled_df.loc[shuffled_df[target]==0].sample(n=L, random_state=42)
    return pd.concat([target_1_df, target_0_df])

```

```python
from sklearn.linear_model import LogisticRegression

logreg = make_pipeline(preprocessing.MinMaxScaler(), LogisticRegression())

#
# Cambiamos el solver, quitando por defecto el cuasi-newton lbfgs que usa el gradiente pero no usa la matriz hessiana
# Como la función no es muy compleja, no habrá problemas de memoria y así podemos probar las penalizaciones l1 y l2
# con norma 1, distancia Manhattan y norma2, euclídea y C, factor inverso a la fuerza de regularización
#

param_grid = {'logisticregression__C': [0.01, 0.1, 1, 2, 10, 100],\
              'logisticregression__penalty':['l1', 'l2'],\
              'logisticregression__solver':['liblinear'],\
              'logisticregression__random_state': [0]}


scoring = {'AUC': 'roc_auc', 'Accuracy': 'accuracy', 'Neg_log_loss': 'neg_log_loss'}

#
# Para el scoring usamos el área bajo la ROC, la exactitud y el log_loss por menos uno.
# En el refit podríamos definir que se ajuste el dataset con la mejor exactitud porque las clases están balanceadas,
# pero utilizamos la AUC para generalizar.

#
# El log_loss penaliza según la probabilidad asignada, la exactitud es menos "escalonada".
# El neg_log_loss es negativo para que cualquier score sea mejor cuanto mayor.
# El neg_log_loss multiplicado por -1 es el log_loss.
#

search = GridSearchCV(logreg, param_grid, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=1),\
                      return_train_score=True, scoring = scoring, refit = 'AUC')

search.fit(X, np.ravel(y))
results = search.cv_results_

print('Parámetros óptimos de la regresión logística con entrenamiento y validación:\n\t' +\
      'C: ' + str(search.best_params_['logisticregression__C']) +\
      '\n\tpenalty: ' + str(search.best_params_['logisticregression__penalty']) +\
      '\n\tsolver: ' + str(search.best_params_['logisticregression__solver']))


mean_train_AUC = results['mean_train_AUC'] [search.best_index_]
std_train_AUC = results['std_train_AUC'] [search.best_index_]
mean_test_AUC = results['mean_test_AUC'] [search.best_index_]
std_test_AUC = results['std_test_AUC'] [search.best_index_]
mean_train_Accuracy = results['mean_train_Accuracy'] [search.best_index_]
std_train_Accuracy = results['std_train_Accuracy'] [search.best_index_]
mean_test_Accuracy = results['mean_test_Accuracy'] [search.best_index_]
std_test_Accuracy = results['std_test_Accuracy'] [search.best_index_]
mean_train_Neg_log_loss = results['mean_train_Neg_log_loss'] [search.best_index_]
std_train_Neg_log_loss = results['std_train_Neg_log_loss'] [search.best_index_]
mean_test_Neg_log_loss = results['mean_test_Neg_log_loss'] [search.best_index_]
std_test_Neg_log_loss = results['std_test_Neg_log_loss'][search.best_index_] 

print('Usamos validación cruzada de 5.\
\nUsamos para el score el AUC, la exactitud y el neg_log_loss.\
\nLas clases están balanceadas, así que la exactitud funciona sin acudir a la precisión, exhaustividad (recall) o al valor-F (F1).\
\nExactitud (accuracy = (TP+TN)/(TP+TN+FP+FN)) de la regresión \
logística en entrenamiento: %0.2f (std %0.2f) \
\nExactitud (accuracy = (TP+TN)/(TP+TN+FP+FN)) de la regresión \
logística en validación: %0.2f (std %0.2f)\
\nAUC de la regresión \
logística en entrenamiento: %0.2f (std %0.2f) \
\nAUC de la regresión \
logística en validación: %0.2f (std %0.2f)\
\nNeg_log_loss de la regresión \
logística en entrenamiento: %0.2f (std %0.2f) \
\nNeg_log_loss de la regresión \
logística en validación: %0.2f (std %0.2f)'\
% (mean_train_Accuracy, std_train_Accuracy, mean_test_Accuracy, std_test_Accuracy,\
   mean_train_AUC, std_train_AUC,mean_test_AUC, std_test_AUC,\
  mean_train_Neg_log_loss, std_train_Neg_log_loss, mean_test_Neg_log_loss, std_test_Neg_log_loss ))

# Usamos test para obtener las métricas de prestaciones
pred_test = search.predict(X_test)
print('Matriz de confusión en test con refit de AUC:')
print(confusion_matrix(y_test,pred_test))
print('Reporte de clasificación en test con refit de AUC:')
print(classification_report(y_test,pred_test))
print('ROC AUC: ' + str(roc_auc_score(y_test, pred_test)))


# Guardamos el search de la regresión logística para tenerlo disponible en otro momento
search_rl = search
```

```python
from sklearn.svm import SVC

# O ponemos la probabilidad a True, o usamos después un CalibratedClassifierCV
clf = make_pipeline(preprocessing.MinMaxScaler(), SVC(probability=True))

#
# Usamos el criterio de divión Gini por defecto.
# Probamos con distintas profundidades, cantidad mínima de muestras por nodo y cantidad de árboles
#

param_grid = {'svc__C':  [0.01, 0.1, 1, 2, 10, 100],\
              'svc__gamma':['scale', 'auto'],\
              'svc__random_state': [0]}

scoring = {'AUC': 'roc_auc', 'Accuracy': 'accuracy', 'Neg_log_loss': 'neg_log_loss'}

#
# Para el scoring usamos el área bajo la ROC, la exactitud y el log_loss por menos uno.
# En el refit podríamos definir que se ajuste con la mejor exactitud,
# pero utilizamos la AUC para generalizar.

#
# El log_loss penaliza según la probabilidad asignada, la exactitud es menos "escalonada".
# El neg_log_loss es negativo para que cualquier score sea mejor cuanto mayor.
# El neg_log_loss multiplicado por -1 es el log_loss.
# Como las clases están balanceadas, usaremos la exactitud en el refit para ajustar todo el dataset con esa
# función de score, aunque probemos con las tres (AUC, exactitud y neg_log_loss)
#

search = GridSearchCV(clf, param_grid, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=1),  return_train_score=True, scoring = scoring, refit = 'AUC')
search.fit(X, np.ravel(y))
results = search.cv_results_

print('Parámetros óptimos de SVM con entrenamiento y validación:\n\t' +\
      'C: ' + str(search.best_params_['svc__C']) +\
      '\n\tgamma: ' + str(search.best_params_['svc__gamma']))


mean_train_AUC = results['mean_train_AUC'] [search.best_index_]
std_train_AUC = results['std_train_AUC'] [search.best_index_]
mean_test_AUC = results['mean_test_AUC'] [search.best_index_]
std_test_AUC = results['std_test_AUC'] [search.best_index_]
mean_train_Accuracy = results['mean_train_Accuracy'] [search.best_index_]
std_train_Accuracy = results['std_train_Accuracy'] [search.best_index_]
mean_test_Accuracy = results['mean_test_Accuracy'] [search.best_index_]
std_test_Accuracy = results['std_test_Accuracy'] [search.best_index_]
mean_train_Neg_log_loss = results['mean_train_Neg_log_loss'] [search.best_index_]
std_train_Neg_log_loss = results['std_train_Neg_log_loss'] [search.best_index_]
mean_test_Neg_log_loss = results['mean_test_Neg_log_loss'] [search.best_index_]
std_test_Neg_log_loss = results['std_test_Neg_log_loss'][search.best_index_] 

print('Usamos validación cruzada de 5.\
\nUsamos para el score el AUC, la exactitud y el neg_log_loss.\
\nLas clases están balanceadas, así que la exactitud funciona sin acudir a la precisión, exhaustividad (recall) o al valor-F (F1).\
\nExactitud (accuracy = (TP+TN)/(TP+TN+FP+FN)) del SVM \
en entrenamiento: %0.2f (std %0.2f) \
\nExactitud (accuracy = (TP+TN)/(TP+TN+FP+FN)) del SVM \
en validación: %0.2f (std %0.2f)\
\nAUC del SVM \
en entrenamiento: %0.2f (std %0.2f) \
\nAUC del SVM \
en validación: %0.2f (std %0.2f)\
\nNeg_log_loss del SVM \
en entrenamiento: %0.2f (std %0.2f) \
\nNeg_log_loss del SVM \
en validación: %0.2f (std %0.2f)'\
% (mean_train_Accuracy, std_train_Accuracy, mean_test_Accuracy, std_test_Accuracy,\
   mean_train_AUC, std_train_AUC,mean_test_AUC, std_test_AUC,\
  mean_train_Neg_log_loss, std_train_Neg_log_loss, mean_test_Neg_log_loss, std_test_Neg_log_loss ))

# Usamos test para obtener las métricas de prestaciones
pred_test = search.predict(X_test)
print('Matriz de confusión en test con refit de AUC:')
print(confusion_matrix(y_test,pred_test))
print('Reporte de clasificación en test con refit de AUC:')
print(classification_report(y_test,pred_test))
print('ROC AUC: ' + str(roc_auc_score(y_test, pred_test)))


# Guardamos el search del bosque para tenerlo disponible en otro momento
search_r4 = search
```
